```python
# setting up parameters for plotting
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Computer Modern Roman'
fontsize_label = 12
fontsize_title = fontsize_label + 1
fontsize_legend = fontsize_label - 4
```

```python
fig, ax = plt.subplots(1,2, figsize=(8,3))
x = np.linspace(0,2,100)
y_test = (x-1.2)**2 + 0.5
y_train = 1/x - 0.5 + 0.01*x
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def mapping(x, alpha=1):
    return (alpha - sigmoid(10*(x-1))) * (x-1)**2
y_train = mapping(x) + 0.2/x
ax[0].plot(x, y_test, label="test error")
ax[0].plot(x, y_train, label="train error")
ax[0].set_ylim([0,2])
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_xlabel("model complexity", fontsize=fontsize_label)
ax[0].legend(fontsize=fontsize_legend)
ax[0].set_title("Bias-Variance Tradeoff", fontsize=fontsize_title)

y_train_2 = mapping(2*x) + 0.3/x
y_test_2 = mapping(x) + 0.5/x - 0.05
ax[1].plot(x, y_test_2, label="test error")
ax[1].plot(x, y_train_2, label="train error")
ax[1].set_ylim([0,2])
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_xlabel("model complexity", fontsize=fontsize_label)
ax[1].legend(fontsize=fontsize_legend)
ax[1].set_title("Modern Observations", fontsize=fontsize_title)
ax[0].axvline(x=1.2, color='black', linestyle='--')
ax[1].axvline(x=1.2, color='black', linestyle='--')
#ax[0].set_ylabel(r'$\text{error}$', fontsize=fontsize_label)
plt.savefig("old_new.pdf", dpi=300, bbox_inches='tight')
plt.show()
```
